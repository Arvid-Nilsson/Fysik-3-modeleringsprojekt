```python
# Importerar bibliotek
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Definerar variabler

# Massan på objektet
mass = 0.002

# Startpunkt i x-led
initial_x = 0

# Startpunkt i y-led
initial_y = 0

# Starthasighet
initial_v = 18

# Tidssteg
delta_t = 0.001

# Radie
r = 0.02

# Startvinkel i grader
alpha = 45

# Luftmotståndskoefficienten
C_D = 0.47

# Tvärsnittsarea
A = r**2 * math.pi

# Luftdensitet
P = 1.2

# Gravitationskonstant
g = 9.82
```

```python
x = initial_x
y = initial_y
v = initial_v
v_x = initial_v * np.cos(np.radians(alpha))
v_y = initial_v * np.sin(np.radians(alpha))
theta = np.radians(alpha)

positions = np.array([[initial_x, initial_y]]) 

k_d = 0.5 * C_D * P * A
```

```python
while y >= 0:
    F_D = k_d * v**2
    F_x = -F_D * np.cos(theta)
    F_y = -mass * g -F_D * np.sin(theta)
    
    a_y = F_y / mass 
    a_x = F_x / mass

    v_x = v * np.cos(theta)
    v_y = v * np.sin(theta)

    v_x = v_x + a_x * delta_t
    v_y = v_y + a_y * delta_t

    x = x + v_x * delta_t
    y = y + v_y * delta_t

    v = np.sqrt(v_x ** 2 + v_y ** 2)
    theta = np.arctan2(v_y, v_x)

    positions = np.append(positions, [[x, y]], axis=0)
```

```python
x_values = positions[:, 0] 
y_values = positions[:, 1]
```

```python
plt.plot(x_values, y_values, label='Approximated Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Kast med luftmotstånd")
plt.legend()
plt.grid(True)
plt.show()
```
