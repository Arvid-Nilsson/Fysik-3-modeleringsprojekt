```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
diameter = 0.02
lDensity = 1420
sDensity = 7870
startV = 0
g = 9.82
stopPosition = 50
deltaT = 0.01


r = 90 * diameter / 2
volume = (4 * math.pi * ((diameter/2) ** 2)) / 3
mass = sDensity * volume
```

```python
def calcFg(tempMass, tempG):
    return tempMass * tempG
def calcFl(tempP, tempV, tempG):
    return tempP * tempV * tempG
def calcFv(tempR, tempV):
    return tempR *tempV
def calcAccel(tempF, tempM):
    return tempF / tempM
```

```python
positionsArray = np.array([0])
timeArray = np.array([0])
velocityArray = np.array([startV])
```

```python
Fg = calcFg(mass, g)
Fl = calcFl(lDensity, volume, g)
Fv = calcFv(r,startV)
v = startV
t = 0

position = 0

while (position <= stopPosition):
    a = calcAccel(Fg-Fl-Fv, mass)
    v = v + a * deltaT
    velocityArray = np.concatenate((velocityArray, np.array([v])))
    position = position + v * deltaT
    positionsArray = np.concatenate((positionsArray, np.array([position])))
    Fv = calcFv(r,v)
    
    t = t + deltaT
    timeArray = np.concatenate((timeArray, np.array([t])))
```

```python
plt.plot(timeArray, positionsArray)
plt.show
```

```python
plt.plot(timeArray, velocityArray)
plt.show
```
