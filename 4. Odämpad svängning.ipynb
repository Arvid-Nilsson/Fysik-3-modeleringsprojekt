```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Mass of obeject
m = 0.3
# Spring constant
k = 3
#Gravitational acceleration
g = 9.82

#starting velocity
v0 = 10

#starting position
y0 = 0

#Steplength 
deltaT = 0.01

#Time of end of simulation
endT = 20
```

```python

def calcForce(tempMass, tempG, tempK, tempY):
    """Function to calculate the force acting uppon the object"""
    return -tempMass*tempG - tempK *tempY

def calcAccel(tempF, tempMass):
    """Function to calcuulate the acceleration acting uppon the object"""
    return tempF / tempMass
```

```python
#Creating arrays for time and position
timeArray = np.array([0])
positionArray = np.array([y0])

#Initializing the current values to starting values
y = y0 
v = v0
t = 0

while (t <= endT):

    #Calculate the acceleration
    a = calcAccel(calcForce(m,g,k,y),m)

    #Calculate the velocity
    v = v + a * deltaT

    #Calculate the position and saving it to array
    y = y + v * deltaT
    positionArray = np.append(positionArray, y)

    #Incrementing the time and adding it to the array
    t += deltaT
    timeArray = np.append(timeArray,t)
```

```python
#Plotting the graphs
plt.plot(timeArray, positionArray)
plt.xlabel("time (s)")
plt.ylabel("y position (m)")
plt.show
```
